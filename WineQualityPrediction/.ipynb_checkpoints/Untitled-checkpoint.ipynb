```python
# Importing necessary libraries
import pandas as pd                     # For data manipulation and analysis
import seaborn as sns                   # For data visualization
import matplotlib.pyplot as plt         # For plotting graphs

# Random Forest model
from sklearn.ensemble import RandomForestClassifier  
 # Support Vector Classifier (SVM model)
from sklearn.svm import SVC                         

# Importing evaluation metrics
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Data preprocessing
# For scaling and encoding
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Model selection utilities
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score

# Magic command for inline plotting in Jupyter Notebook
%matplotlib inline
```

```python
df=pd.read_csv("data.csv")
```

```python
df
```

```python
fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'fixed acidity', data = df)
```

```python
bins=(2,6.5,8)
group_names=['bad','good']
df['quality']=pd.cut(df['quality'],bins=bins,labels=group_names)
```

```python
df['quality']
```

```python
label=LabelEncoder()
```

```python
df['quality']=label.fit_transform(df['quality'])
```

```python
x=df.drop('quality' ,axis=1)
y=df['quality']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
Rfc=RandomForestClassifier(n_estimators=200)
Rfc.fit(x_train,y_train)
predict_rfc=Rfc.predict(x_test)
```

```python
print("Accuaracy score =",accuracy_score(y_test, predict_rfc))
print(classification_report(y_test, predict_rfc))
```

```python
#use support vector machine imporce accuary
svc = SVC()
svc.fit(x_train, y_train)
pred_svc = svc.predict(x_test)
```

```python
print("Accuaracy score =",accuracy_score(y_test, predict_rfc))
print(classification_report(y_test, predict_rfc))
```

```python
param = {
    'C': [0.1,0.8,0.9,1,1.1,1.2,1.3,1.4],
    'kernel':['linear', 'rbf'],
    'gamma' :[0.1,0.8,0.9,1,1.1,1.2,1.3,1.4]
}
grid_svc = GridSearchCV(svc,param, cv=10, verbose=2)
```

```python
grid_svc.fit(x_train, y_train)
```

```python
pred = grid_svc.predict(x_test)
print("Accuaracy score =", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

```python
rfc_eval = cross_val_score(estimator = Rfc, X = x_train, y = y_train, cv = 10, verbose=2)
rfc_eval.mean()
```
